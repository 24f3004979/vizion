## Problem-4

Center the the dataset $\mathbf{X}$ around its mean. Call this centered dataset $\mathbf{X}^{\prime}$.

<u>Conversion</u>

Enter the following quantity as your answer correct to two decimal places:

$$
\cfrac{1}{n \cdot d} \sqrt{\sum \limits_{i = 1}^{n} \sum \limits_{j = 1}^{d} (X_{ij}^{\prime})^2}
$$

```python
import math

# Calculate the mean of each feature (pixel) across all samples
X_mean = np.mean(X, axis=0)

# Center the dataset by subtracting the feature-wise mean
X_prime = X - X_mean

# Applying given formula for Problem 4 conversion (using n=100, d=784)
n_samples = X.shape[0] # Number of samples
n_features = X.shape[1] # Number of features

root_target = np.sum(X_prime**2)

final = (1 / (n_samples * n_features)) * math.sqrt(root_target)
print(f"Problem 4 conversion quantity: {final}")
```

## Problem-6

Compute the first and second principal components of the dataset, $\mathbf{w}_1$ and $\mathbf{w}_2$.

<u>Conversion</u>

Compute:
$$
\mathbf{w} = \mathbf{w}_1 + \mathbf{w}_2
$$

Enter the following quantity as your answer correct to two decimal places:

$$
\left| \sum \limits_{i = 1}^{d} w_i \right|
$$

```python
import numpy as np

# Compute the covariance matrix C of the centered dataset X_prime
# Setting rowvar=False means that columns are variables (features)
# so C will be (n_features, n_features) = (784, 784)
C = np.cov(X_prime, rowvar=False)

print(f"Shape of covariance matrix C: {C.shape}")
```

```python
import numpy as np

# Compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(C)

# Take the real part to handle potential complex values due to numerical precision
eigenvalues = np.real(eigenvalues)
eigenvectors = np.real(eigenvectors)

# Sort eigenvalues and eigenvectors in descending order
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]

k = 2

s = 0
for i in eigenvectors:
  w = i[0] + i[1]
  s += w
print(s)
```

## Problem-7

Visualize the first principal component as an image.

```python
import matplotlib.pyplot as plt

# The first principal component is the first column of the sorted eigenvectors
# After fixing the covariance matrix and re-running, eigenvectors[:, 0] will be 784-dimensional
first_principal_component = eigenvectors[:, 0]
print(f"First principal component shape: {first_principal_component.shape}")

# Reshape the component into a 28x28 image (28*28 = 784)
image_shape = (28, 28)
first_pc_image = first_principal_component.reshape(image_shape)

# Display the image
plt.imshow(first_pc_image, cmap='gray')
plt.title('First Principal Component')
plt.colorbar()
plt.show()
```

## Problem-8

Find the smallest value of $k$ for which the first $k$ principal components capture at least $95\%$ of the variance in the data.

```python
# Calculate total variance
total_variance = np.sum(eigenvalues)

# Calculate explained variance ratio for each component
explained_variance_ratio = eigenvalues / total_variance

# Calculate cumulative explained variance ratio
cumulative_explained_variance_ratio = np.cumsum(explained_variance_ratio)

# Find the smallest k for 95% variance
k_95_percent = np.argmax(cumulative_explained_variance_ratio >= 0.95) + 1

print(f"The smallest k for 95% variance is: {k_95_percent}")
```

## Problem-9

Given a `test_image`, visualize the proxies  by reconstructing it using the top $k$ principal components. Consider four values of $k$; values of $k$ for which the top-$k$ principal components explain:

- 20% of the variance
- 50% of the variance
- 80% of the variance
- 95% of the variance



```python
import matplotlib.pyplot as plt

# Defined variance thresholds for Problem 9
variance_thresholds = [0.20, 0.50, 0.80, 0.95]

# Center the test_image using the previously computed X_mean
test_image_centered = test_image - X_mean

# Calculate total variance from eigenvalues
total_variance = np.sum(eigenvalues)
explained_variance_ratio = eigenvalues / total_variance
cumulative_explained_variance_ratio = np.cumsum(explained_variance_ratio)

plt.figure(figsize=(15, 5))

for i, threshold in enumerate(variance_thresholds):
    # Find the smallest k for the current variance threshold
    k_value = np.argmax(cumulative_explained_variance_ratio >= threshold) + 1

    # Select the top k eigenvectors (principal components)
    top_k_eigenvectors = eigenvectors[:, :k_value]

    # Project the centered test image onto the top k principal components
    projected_test_image = test_image_centered @ top_k_eigenvectors

    # Reconstruct the image from the projected components
    reconstructed_image_centered = projected_test_image @ top_k_eigenvectors.T

    # Add back the mean to get the reconstructed image in the original scale
    reconstructed_image = reconstructed_image_centered + X_mean

    # Reshape the reconstructed image to 28x28 for visualization
    reconstructed_image_2d = reconstructed_image.reshape(28, 28)

    # Plot the reconstructed image
    plt.subplot(1, len(variance_thresholds), i + 1)
    plt.imshow(reconstructed_image_2d, cmap='gray')
    plt.title(f'k={k_value} ({int(threshold*100)}% var)')
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
photo_link = '/home/madhav/Downloads/p.jpg'

# Simple visual drawing shot with pca 
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

def mathematically_sketch_image(image_path):
    # 1. Load the image and convert it to grayscale (a 2D mathematical matrix)
    try:
        img = Image.open(image_path).convert('L')
    except FileNotFoundError:
        print(f"Error: Could not find '{image_path}'. Please check the path.")
        return
        
    img_matrix = np.array(img)

    # 2. Perform Singular Value Decomposition (SVD)
    # This extracts the eigenvectors (U, Vt) and their weights (S)
    print("Calculating eigenvectors...")
    U, S, Vt = np.linalg.svd(img_matrix, full_matrices=False)

    # 3. Define how many components we want to use for our "sketches"
    # The maximum number of components is the smaller dimension of the image
    max_components = min(img_matrix.shape)
    components_to_test = [5, 20, 50, max_components]
    
    # 4. Set up the visualization
    fig, axes = plt.subplots(1, 4, figsize=(16, 5))
    fig.suptitle("Image Reconstruction via Eigenvectors (PCA/SVD)", fontsize=16)

    for i, k in enumerate(components_to_test):
        # Reconstruct the matrix using only the top 'k' components
        # Math: U_k * Sigma_k * Vt_k
        reconstructed = np.dot(U[:, :k], np.dot(np.diag(S[:k]), Vt[:k, :]))
        
        # Display the math as an image
        axes[i].imshow(reconstructed, cmap='gray')
        if k == max_components:
            axes[i].set_title(f'Exact Picture ({k} PCs)')
        else:
            axes[i].set_title(f'Sketch with {k} PCs')
        axes[i].axis('off')

    plt.tight_layout()
    plt.show()

mathematically_sketch_image(photo_link)
```
