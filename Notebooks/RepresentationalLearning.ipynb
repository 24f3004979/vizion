```python
from sklearn.datasets import load_iris

iris = load_iris(as_frame=True)
print(iris.keys())

import seaborn as sns

# Rename classes using the iris target names
iris.frame["target"] = iris.target_names[iris.target]
_ = sns.pairplot(iris.frame, hue="target")
```

## Observations
With using PCA we compressed the features which were given for all three flowers type and expressed those features through 3 fundamental vectors,
Thus we have used less number of features to express the differences and structure of those features at 3d plot

We simply taken down too many features and compressed their essense with PCA into 3 dimension and expressed the all 3 given units with those dimensions, 

Doing Representatinal Learning about dataset and reducing total space required to store them :)
