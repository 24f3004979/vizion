```python
import matplotlib.pyplot as plt
import numpy
import pandas as pd

# loading Dataset
housing = pd.read_csv('/content/sample_data/california_housing_test.csv')

# Making simple numerical extraction with dataset
columns = [i for i in housing]
print(f"data types : {columns}")

'''
Geting information about datatype of each column for initial filteration and pivot generations
Information about Pivot
  making basic column data analysis for the final pivot recommendations
'''

type_listing = [ housing[colm].dtype for colm in columns]

print(f'data set check : {type_listing}')

housing.head()
```

```python
from matplotlib.cbook import file_requires_unicode
#@title Plot testing

freq_total_rooms = housing['total_rooms'].value_counts()

# Series dataset with frequency :: Extract with dict way
X = []
Y = []
for elem in freq_total_rooms.keys():
  X.append(elem)
  Y.append(freq_total_rooms[elem])

print(f"X is Number of such house {len(X), X}")
print(f"Y is frequency of such : {len(Y), Y}")
plt.scatter(Y, X)
plt.xlabel('Room count')
plt.ylabel('Frequency in dataset')
plt.title(" Plot testing ")
```

## Correlation Matrix Analysis
Finding dependenncy of vector with finding, directly proportional and indirectly proportional units

**How to make correlation Matrix**
1. Find mean for both vectors
2. Sum of all units subtracted with their mean multiplied
3. divided by standard deviation of both vectors

$$
centric = {\sum}(x_i - \bar x)(y_i - \bar y)
$$

$$
deviation ⏫ d = {\sum}(x_i - \bar x)^2{\sum}(y_i - \bar y)^2
$$

$$
SD = \sqrt D
$$

$ D $ Serves as the main form for expression of surface level linear relations between variables and a gate to make simple visualization of those elements.

Correlations wont give us Non linear relations between two variable thus we have to sure that we are just geting the linear estimations about their relation enough for basic view 🔰
